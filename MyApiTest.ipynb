```python
# Set Varaiable 

apiSite = 'https://forkify-api.herokuapp.com/api/v2/recipes?search=pizza'
```

```python
#Koda for
import requests
import json

# Make a GET request to the API
response = requests.get('https://forkify-api.herokuapp.com/api/v2/recipes?search=pizza')

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON response
    data = response.json()

    # Print the data in a nice format
    rprint(json.dumps(data, indent=4))
    pprint(json.dumps(data, indent=4))
else:
    print(f"Request failed with status code {response.status_code}")
```

```python
# same above with try catch block
import requests
import json

try:
    # Make a GET request to the API
    response = requests.get('https://forkify-api.herokuapp.com/api/v2/recipes?search=pizza')

    # Raise an exception if the request was not successful
    response.raise_for_status()

    # Parse the JSON response
    data = response.json()

    # Print the data in a nice format
    rprint(json.dumps(data.keys, indent=4))

except requests.exceptions.RequestException as err:
    rprint(f"Request failed: {err}")
except json.JSONDecodeError as err:
    rprint(f"Failed to parse JSON response: {err}")
```

```python
import requests
import pandas as pd
from tabulate import tabulate

apiSite = 'https://forkify-api.herokuapp.com/api/v2/recipes?search=pizza'

try:
    # Make a GET request to the API
    response = requests.get(apiSite)

    # Raise an exception if the request was not successful
    response.raise_for_status()

    # Parse the JSON response
    data = response.json()

    # Convert the data to a DataFrame
    df = pd.DataFrame(data)

    # Print the DataFrame in a nice table format
    rprint(tabulate(df, headers='keys', tablefmt='psql'))

except requests.exceptions.RequestException as err:
    print(f"Request failed: {err}")
except json.JSONDecodeError as err:
    print(f"Failed to parse JSON response: {err}")
```
